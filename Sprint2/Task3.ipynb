<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Этот проект на C# демонстрирует реализацию системы для работы с книгами различных типов (учебники, художественная литература, научная литература) с использованием интерфейсов и классов. Основная цель — симулировать действия по работе с книгами, такими как чтение, взятие на время, возврат, исследование и получение информации о книге.

Ключевые элементы проекта:
Интерфейсы:

IBorrowable — предоставляет методы для взятия и возврата книг.
IReadable — предоставляет метод для чтения книг.
Классы:

Book — базовый класс, представляющий книгу с основными свойствами (название, автор, год публикации, издательство и доступность). Он включает методы для проверки доступности книги, резервации и возврата.
TextBook — подкласс Book, реализующий интерфейсы IBorrowable и IReadable, добавляющий свойства "Предмет" и "Уровень образования", а также методы для изучения и чтения учебника.
Fiction — подкласс Book, реализующий IBorrowable, с добавлением жанра и главного персонажа. Методы позволяют брать книгу и описывать главного персонажа.
ScientificLiterature — подкласс Book, реализующий IReadable, добавляющий свойства "Область науки" и "Рецензируемость", а также методы для чтения и проведения исследований.
Основная логика: В основной программе создаются объекты для каждого типа книг:

Учебник по математике.
Художественная книга в жанре "Тайна".
Научная литература по квантовой механике.
Пример использования включает:

Чтение книг.
Взятие учебника на время и его возврат.
Описание главного персонажа художественной книги.
Проведение исследований по научной литературе.

```
using System;


public interface IBorrowable
{
    string Borrow();
    string Return();
}


public interface IReadable
{
    string Read();
}


public class Book
{
    
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }
    public string Publisher { get; set; }
    public int Pages { get; set; }
    public bool IsAvailable { get; set; }

    
    public Book(string title, string author, int yearOfPublication, string publisher, int pages)
    {
        Title = title;
        Author = author;
        YearOfPublication = yearOfPublication;
        Publisher = publisher;
        Pages = pages;
        IsAvailable = true;
    }

   
    public virtual string GetInfo()
    {
        return $"Название: {Title}\nАвтор: {Author}\nГод публикации: {YearOfPublication}\nПублицации: {Publisher}\nСтраницы: {Pages}";
    }

    public string Reserve()
    {
        if (IsAvailable)
        {
            IsAvailable = false;
            return "Книга зарезервирована.";
        }
        return "Книга уже зарезервирована.";
    }

    public string ReturnBook()
    {
        IsAvailable = true;
        return "Книга возвращена.";
    }

    public string CheckAvailability()
    {
        return IsAvailable ? "Книга доступна." : "Книга недоступна.";
    }
}


public class TextBook : Book, IBorrowable, IReadable
{
    public string Subject { get; set; }
    public string GradeLevel { get; set; }

    
    public TextBook(string title, string author, int yearOfPublication, string publisher, int pages, string subject, string gradeLevel)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        Subject = subject;
        GradeLevel = gradeLevel;
    }

    
    public string Read()
    {
        return $"Вы читаете учебник по {Subject}.";
    }

    public string Borrow()
    {
        return "Вы взяли учебник.";
    }

    public string Return()
    {
        return "Вы вернули учебник.";
    }

    
    public string Study()
    {
        return $"Вы изучаете {Subject} на уровне {GradeLevel}.";
    }
}


public class Fiction : Book, IBorrowable
{
    public string Genre { get; set; }
    public string MainCharacter { get; set; }

    
    public Fiction(string title, string author, int yearOfPublication, string publisher, int pages, string genre, string mainCharacter)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        Genre = genre;
        MainCharacter = mainCharacter;
    }

    
    public string Borrow()
    {
        return $"Вы взяли книгу жанра {Genre}.";
    }

    public string Return()
    {
        return "Вы вернули художественную книгу.";
    }

    
    public string DescribeMainCharacter()
    {
        return $"Главный персонаж этой книги — {MainCharacter}.";
    }
}

public class ScientificLiterature : Book, IReadable
{
    public string FieldOfScience { get; set; }
    public bool IsPeerReviewed { get; set; }

    
    public ScientificLiterature(string title, string author, int yearOfPublication, string publisher, int pages, string fieldOfScience, bool isPeerReviewed)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        FieldOfScience = fieldOfScience;
        IsPeerReviewed = isPeerReviewed;
    }

    
    public override string GetInfo()
    {
        return $"{base.GetInfo()}\nОбласть науки: {FieldOfScience}\nРецензируемый: {(IsPeerReviewed ? "ДА" : "НЕТ")}";
    }

    
    public string Read()
    {
        return $"Вы читаете научную литературу по {FieldOfScience}.";
    }

    
    public string ConductResearch()
    {
        return $"Вы проводите исследование в области {FieldOfScience}.";
    }
}

        TextBook mathBook = new TextBook("Математика для инженера", "Джон", 2020, "Научный пресс", 350, "Математика", "Студент");
        Fiction novel = new Fiction("Тайна старого дома", "Элис Джонсон", 2018, "Новая пресса", 250, "Тайна", "Детектив Смит");
        ScientificLiterature scienceBook = new ScientificLiterature("Квантовая механика", "Альберт Эйнштейн", 1925, "Физический пресс", 400, "Физике", true);

        
        Console.WriteLine(mathBook.Study());                 
        Console.WriteLine(novel.DescribeMainCharacter());    
        Console.WriteLine(scienceBook.ConductResearch());    

       
        Console.WriteLine(mathBook.Read());                 
        Console.WriteLine(scienceBook.Read());              

        
        Console.WriteLine(mathBook.Borrow());               
        Console.WriteLine(mathBook.Return());               

        
        Console.WriteLine(novel.CheckAvailability());        
```
