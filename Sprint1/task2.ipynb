Задание:
Ниже в блоке по примеру создайте базовый класс Animal и производные классы (2-3 например Dog, Cat и так далее) реализуйте структуру и объявление класса, включая свойства, геттеры и сеттеры, а также статические и обычными атрибуты и метода класса с реализацией взаимодействия между объектами.

```python
using System;

// Базовый класс Animal
public class Animal
{
    private string _name;
    private int _age;

    // Свойство Name с геттером и сеттером
    public string Name
    {
        get { return _name; }
        set { _name = value; }
    }

    // Свойство Age с геттером и сеттером
    public int Age
    {
        get { return _age; }
        set
        {
            if (value >= 0)
                _age = value;
            else
                throw new ArgumentOutOfRangeException("Возраст не может быть отрицательным!");
        }
    }

    // Статическое поле для подсчета количества животных
    public static int TotalAnimals { get; private set; } = 0;

    // Конструктор
    public Animal(string name, int age)
    {
        _name = name;
        _age = age;
        TotalAnimals++;
    }

    // Метод для получения информации об Animal
    public virtual void GetInfo()
    {
        Console.WriteLine($"Name: {Name}, Age: {Age}");
    }

    // Статический метод для отображения общего количества животных
    public static void DisplayTotalAnimals()
    {
        Console.WriteLine($"Total number of animals: {TotalAnimals}");
    }
}

// Производный класс Dog
public class Dog : Animal
{
    private string _breed;

    // Свойство Breed с геттером и сеттером
    public string Breed
    {
        get { return _breed; }
        set { _breed = value; }
    }

    // Конструктор
    public Dog(string name, int age, string breed) : base(name, age)
    {
        _breed = breed;
    }

    // Переопределенный метод GetInfo
    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Breed: {Breed}");
    }

    // Метод для собачьего лая
    public void Bark()
    {
        Console.WriteLine($"{Name} is barking!");
    }
}

// Производный класс Cat
public class Cat : Animal
{
    private string _color;

    // Свойство Color с геттером и сеттером
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Конструктор
    public Cat(string name, int age, string color) : base(name, age)
    {
        _color = color;
    }

    // Переопределенный метод GetInfo
    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Color: {Color}");
    }

    // Метод для кошачьего мяуканья
    public void Meow()
    {
        Console.WriteLine($"{Name} is meowing!");
    }
}

        // Создаем объекты Dog и Cat
        Dog myDog = new Dog("Buddy", 3, "Labrador");
        Cat myCat = new Cat("Whiskers", 2, "Tabby");

        // Вывод информации о животных
        myDog.GetInfo();
        myCat.GetInfo();

        // Взаимодействие между объектами
        myDog.Bark();
        myCat.Meow();

        // Отображаем общее количество животных
        Animal.DisplayTotalAnimals();
```
